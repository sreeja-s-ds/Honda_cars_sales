```python
import pandas as pd
```

```python
df=pd.read_csv('/content/honda_car_selling.csv')
```

```python
df
```

```python
df.info()
```

```python
for col in ['Price', 'kms Driven']:
    df[col] = df[col].str.replace(',', '').str.strip()  # remove commas and extra spaces

    if col == 'Price':
        df[col] = df[col].str.replace('Lakh', '', regex=False).astype(float) * 100000
    elif col == 'kms Driven':
        df[col] = df[col].str.replace('kms', '', regex=False).astype(int)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder


# Initialize LabelEncoder
le = LabelEncoder()

# Columns to encode
cols_to_encode = ['Fuel Type', 'Suspension', 'Car Model']

# Encode each column using a for loop
for col in cols_to_encode:
    df[col] = le.fit_transform(df[col])
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
from scipy import stats

# Make a copy of the DataFrame
df_stat = df.copy()
```

```python
qa_results = []

# 1. T-test: Automatic vs Manual car prices
auto_prices = df_stat[df_stat['Suspension'] == 0]['Price']
manual_prices = df_stat[df_stat['Suspension'] == 1]['Price']
t_stat, p_val = stats.ttest_ind(auto_prices, manual_prices, equal_var=False)
qa_results.append(("Automatic vs Manual car prices", p_val))
print("T-test results:")
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")
```

```python
# Assuming Petrol = 2 and Diesel = 0 from earlier encoding
petrol_kms = df_stat[df_stat['Fuel Type'] == 2]['kms Driven']
diesel_kms = df_stat[df_stat['Fuel Type'] == 0]['kms Driven']

# T-test
t_stat, p_val = stats.ttest_ind(petrol_kms, diesel_kms, equal_var=False, nan_policy='omit')

# Append result
qa_results.append(("Petrol vs Diesel kilometers driven", p_val))
print("\nT-test results:")
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")

# 2. T-test: Petrol vs Diesel kms driven
```

```python
# 3. Chi-Square: Fuel Type vs Suspension
contingency1 = pd.crosstab(df_stat['Fuel Type'], df_stat['Suspension'])
chi2, p_val, _, _ = stats.chi2_contingency(contingency1)
qa_results.append(("Fuel Type vs Suspension", p_val))
print("\nChi-Square results:")
print(f"Chi-Square statistic: {chi2}")
print(f"P-value: {p_val}")
```

```python
# 4. ANOVA: Price across fuel types
groups = [df_stat[df_stat['Fuel Type'] == i]['Price'] for i in df_stat['Fuel Type'].unique()]
f_stat, p_val = stats.f_oneway(*groups)
qa_results.append(("Price across Fuel Types (ANOVA)", p_val))
print("\nANOVA results:")
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_val}")
```

```python
# 5. Chi-Square: Car Model vs Suspension
contingency2 = pd.crosstab(df_stat['Car Model'], df_stat['Suspension'])
chi2, p_val, _, _ = stats.chi2_contingency(contingency2)
qa_results.append(("Car Model vs Suspension", p_val))
print("\nChi-Square results:")
print(f"Chi-Square statistic: {chi2}")
print(f"P-value: {p_val}")
```

```python
# 6. T-test: Older (<2015) vs Newer (>=2015) car prices
old_cars = df_stat[df_stat['Year'] < 2015]['Price']
new_cars = df_stat[df_stat['Year'] >= 2015]['Price']
t_stat, p_val = stats.ttest_ind(old_cars, new_cars, equal_var=False)
qa_results.append(("Old vs New car prices", p_val))
print("\nT-test results:")
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")
```

```python
# 7. ANOVA: Kilometers driven across car years
year_groups = [df_stat[df_stat['Year'] == y]['kms Driven'] for y in df_stat['Year'].unique() if len(df_stat[df_stat['Year'] == y]) > 5]
f_stat, p_val = stats.f_oneway(*year_groups)
qa_results.append(("KMs Driven across Years (ANOVA)", p_val))
print("\nANOVA results:")
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_val}")
```

```python
# 8. Chi-Square: Car Model vs Fuel Type
contingency3 = pd.crosstab(df_stat['Car Model'], df_stat['Fuel Type'])
chi2, p_val, _, _ = stats.chi2_contingency(contingency3)
qa_results.append(("Car Model vs Fuel Type", p_val))
print("\nChi-Square results:")
print(f"Chi-Square statistic: {chi2}")
print(f"P-value: {p_val}")
```

```python
# 9. ANOVA: Price across top 3 car models
top3_models = df_stat['Car Model'].value_counts().nlargest(3).index
model_groups = [df_stat[df_stat['Car Model'] == m]['Price'] for m in top3_models]
f_stat, p_val = stats.f_oneway(*model_groups)
qa_results.append(("Price across Top 3 Car Models (ANOVA)", p_val))
print("\nANOVA results:")
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_val}")
```

```python
# 10. T-test: Price for cars above vs below 50,000 kms
above_50k = df_stat[df_stat['kms Driven'] > 50000]['Price']
below_50k = df_stat[df_stat['kms Driven'] <= 50000]['Price']
t_stat, p_val = stats.ttest_ind(above_50k, below_50k, equal_var=False)
qa_results.append(("Price for >50K vs <=50K km", p_val))
print("\nT-test results:")
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")
```

```python
# Print results
for i, (question, p_val) in enumerate(qa_results, 1):
    result = "Yes" if p_val < 0.05 else "No"
    print(f"{i}. {question}")
    print(f"   p-value: {p_val:.4f} ➤ Significant? {result}\n")
```
